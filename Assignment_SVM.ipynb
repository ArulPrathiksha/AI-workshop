```python
dataset=pd.read_csv("/content/drive/My Drive/ML/Machine Learning-Regression/Assignment/insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children',  'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[["charges"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel="linear", C=100, gamma="scale")
regressor.fit(X_train, y_train)


```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python

y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="/content/drive/My Drive/ML/Machine Learning-Regression/Assignment/finalized_model_Mul_linear1.sav"
pickle.dump(regressor,open(filename,'wb'))
```
